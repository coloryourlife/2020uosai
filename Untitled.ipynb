```python
x = np.array([[0.16,0.93,0.94],[-0.14,0.93,0.94],[-0.44,0.93,0.64],[-0.74,0.63,0.64]])
```

```python
a = [n for n in range(3)]
list = {}
index = 0;
for i in x:
    b = [(-i[1]/i[2])*b-i[0]/i[2] for b in range(3)]
    index +=1
    plt.scatter(0,0)
    plt.scatter(0,1)
    plt.scatter(1,0)
    plt.scatter(1,1)
    plt.title("w0 : "+ str(i[0])+ " w1 : " + str(i[1]) + " w2 : "+str(i[2]))
    plt.plot(a,b)
    plt.show()
```
