```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
m = [0,0,1,1]
n = [0,1,0,1]
x = np.arange(0,2)
y = [0.5/0.03*b + 0.03 for b in range(0,2)]
a = np.arange(0,2)
b = [0.35/0.64*b + 0.19 for b in range(0,2)]
plt.plot(x,y)
plt.plot(a,b)
plt.scatter(m,n)
plt.show()
```

```python
y = [3.87/3.86*b + 0.03 for b in range(0,2)]
b = [5.99/5.94*b + 0.19 for b in range(0,2)]
plt.plot(x,y)
plt.plot(a,b)
plt.scatter(m,n)
plt.show()
```
