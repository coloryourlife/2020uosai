```python
def sigmoid(x):
    return 1 / (1+np.exp(-x))
def forward(w1,w2,bias):
    li = []
    li.append(sigmoid(bias))
    li.append(sigmoid(w2 + bias))
    li.append(sigmoid(w1 + bias))
    li.append(sigmoid(w1 + w2 + bias))
    return li

def dforward(w1,w2,bias):
    li = []
    li.append(sigmoid(bias))
    li.append(sigmoid(w2+bias))
    li.append(sigmoid(w1+bias))
    li.append(sigmoid(w1+w2+bias))
    li.append(sigmoid(0.5*w1+w2+bias))
    li.append(sigmoid(w1+0.5*w2+bias))
    li.append(sigmoid(0.5*w2 + bias))
    li.append(sigmoid(0.5*w1+bias))
    li.append(sigmoid(0.5*w1+0.5*w2+bias))
    return li
        
    
```
